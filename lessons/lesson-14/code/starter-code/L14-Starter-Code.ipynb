```python
import pandas as pd
import json

# this imports the same dataset we used in previous class - lesson 13

data = pd.read_csv("../../../lesson-13/code/dataset/stumbleupon.tsv", sep='\t')
data['title'] = data.boilerplate.map(lambda x: json.loads(x).get('title', ''))
data['body'] = data.boilerplate.map(lambda x: json.loads(x).get('body', ''))
data.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bodies = data['body'].dropna().copy()
vectorizer = CountVectorizer(max_features=1000, binary=False, stop_words='english')

# so the max features ensures we only pick the 1000 most common used words
```

```python
# X = vectorizer.fit(bodies)
# X = vectorizer.transform(bodies)

# alternative command combining both is:

X = vectorizer.fit_transform(bodies)
```

```python
print X

# This print shows for each record the words that occur in that record 
```

```python
id2word = dict(enumerate(vectorizer.get_feature_names()))
```

```python
id2word
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.matutils import Sparse2Corpus

# first we convert our matrix into gensim's format

corpus = Sparse2Corpus(X, documents_columns = False)

# then we fit an LDA model asking to find 15 topics

lda_model = LdaModel(corpus=corpus, id2word=id2word, num_topics=15)

# the max_topics = 15 above specifies that the 1,000 words from above will be grouped into 15 groups
```

```python
# Now print the 15 identified topics
# This will list the words that are in the "topic" and the likelyhood of the word in that topic

# This was clustering so you still need to determine which "label" should go with each of
# these 15 clusters. This can be very difficult!

# enumerate gives access to item and the index of item in matrix at same time; ti stands 
# for topic index


num_topics = 15
num_words_per_topic = 5
for ti, topic in enumerate(lda_model.show_topics(num_topics = num_topics, num_words = num_words_per_topic)):
    print ("Topic: %d" % (ti))
    print(topic)
    print()
    
    
# the number of topics (in this case 15) is a sort of trial and error you need to determine based on
# reviewing the clusters using your domain knowledge (rather subjective process)

```
